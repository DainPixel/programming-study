실습 4: 위치, 크기가 서로 다른 100개의 점 만들기(투명도 설정)
 - alpha 속성으로 투명도를 조절
 - alpha 속성값의 범위: 0 ~ 1
-  0에 가까울수록 투명, 1에 가까울수록 불투명

matplotlib 라이브러리에서 사용할 수 있는 다양한 컬러맵

image.png

```python
#위치, 크기가 서로 다른 100개의 점 만들기(투명도 설정)
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
    
plt.scatter(x, y, s = size, c = size, cmap='jet', alpha =0.7)
plt.colorbar()
plt.show()
```

실습 5: 남녀 연령별 인구 데이터 scatter() 함수로 표현하기

```python

#남녀 연령별 인구 데이터 scatter() 함수로 표현하기

import csv
fp = open(directory_path + 'gender.csv', 'r', encoding='cp949')
data = csv.reader(fp)

m = []
f = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data :
    if name in row[0] :
        for i in range(3,104) :
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break
fp.close()

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.scatter(m, f)
plt.show()
```

```python
#y = x 형태의 추세선 추가하기

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.scatter(m, f, c =range(101), alpha=0.5, cmap='jet')
plt.colorbar()
#추세선 추가
plt.plot(range(max(m)),range(max(m)), 'g')
plt.show()
```

연령대별 성별 비율 산점도로 표현하기(원의 크기는 인구수의 제곱근)

```python
#연령대별 성별 비율 산점도로 표현하기(원의 크기는 인구수의 제곱근)

import csv
import math
fp = open(directory_path + 'gender.csv', 'r', encoding='cp949')
data = csv.reader(fp)
m = []
f = []
size = []
name = input('궁금한 동네를 입력해주세요 : ')
for row in data :
    if name in row[0] :
        for i in range(3,104) :
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i]) + int(row[i+103])))
        break
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')
plt.figure(figsize = (10,5), dpi=300)
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m, f, s = size, c = range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')
plt.xlabel('남성 인구 수')
plt.ylabel('여성 인구 수')
plt.show()
fp.close()
```
