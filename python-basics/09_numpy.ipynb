실습1: 리스트를 구현한 프로그램을 numpy를 사용하여 작성하기

```python
import matplotlib.pyplot as plt
t = []
p2 = []
p3 = []
for i in range(0, 50, 2) :
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

 - arrange 함수: 연속된 값이 들어가있는 numpy array를 생성

  - 파라미터 1개: 0에서 파라미터 -1 까지 연속된 숫자를 가진 numpy array 만듬

  - 파라미터 3개: 첫 번째 파라미터가 시작하는 숫자, 두 번째 파라미터 -1이 끝나는 숫자, 세 번째 파라미터가 간격을 나타냄

- 연속된 실수의 numpy array 만들기

 - linspace 함수: linspace(start, stop, num)
   - 실수를 생성
   - start: 시작 
   - stop: 끝
   - num: start부터 stop까지의 간격의 수

```python
#연속된 실수의 numpy array 만들기

a = np.arange(1, 2, 0.1) # 0부터 2까지 연속된 숫자 생성
b = np.linspace(1, 2, 11) #1부터 2를 11개개의 구간으로 나눈 실수 생성

print(a)
print(b)
```

실습8: Mask 적용하기

```python
#Mask 적용하기(기본 데이터 만들기)
import numpy as np
a = np.arange(-5, 5)
print(a)

#Mask 적용하기(마스크 만들기)
print("Mask 적용하기(마스크 만들기)")
print(a < 0)
print(a[a < 0])
mask1 = abs(a) > 3
print(a[mask1])


#Mask 적용하기(마스크 연결해서 사용하기
mask1 = abs(a) > 3
mask2 = abs(a) % 2 ==0
print(mask1)
print(mask2)
print(a[mask1 + mask2]) # 둘 중 하나의 조건이라도 참일 경우
print(a[mask1 * mask2]) # 두 가지 조건이 모두 참일 경우
```

실습9 : numpy 라이브러리를 사용하여 재미있는 버블 차트 그리기

```python
#numpy 라이브러리를 사용하여 재미있는 버블 차트 그리기

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)          # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)          # 1000개의 랜덤 값 추출
mask1 = abs(x) >50          # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) >50          # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
x = x[mask1 + mask2]        # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2]        # mask1과 mask2 중 하나라도 만족하는 값 저장

num = len(x)
size = np.random.rand(num)*100
plt.scatter(x, y, s = size, c = x, cmap = 'jet', alpha = 0.7)
plt.colorbar()
plt.show()
```
