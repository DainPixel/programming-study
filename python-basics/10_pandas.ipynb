- A열의 값을 B열의 값으로 나운 결과로 D열 만들기

```python
# A열의 값을 B열의 값으로 나운 결과로 D열 만들기
import numpy as np
'''
code 추가
'''
index = pd.date_range('1/1/2000', periods =8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B']  # A열의 값을 B열의 값으로 나눈 값을 D열에 저장
df.head()
```

- 행 우선 계산 값을 E 열에 저장

```python
'''
code 추가
'''
df['E'] = np.sum(df, axis = 1)  #열 우선 계산 값을 E 열에 저장
df.head()
```

- A열의 데이터를 기준으로 열 우선 계산

```python
'''
code 추가
'''
df = df.sub(df['A'], axis =0)  
df.head()
```

- C열의 데이터를 기준으로 열 우선 계산
 - 모든 값을  C열의 값으로 나눈 후 csv 파일로 저장하기

```python
df = df.div(df['C'], axis = 0)   #C열의 데이터를 기준으로 행 우선 계산
df.to_csv(directory_path+'test.csv')
df.head()
```

```python
import pandas as pd

df = pd.read_csv(directory_path + 'age_v2.csv', encoding='cp949', index_col = 0)
'''
code 추가
'''
df = df.div(df['총인구수'], axis =0)     # 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수']  # 총인구수, 연령구간인구수 열 삭제
```

```python
name = input('원하는 지역의 이름을 입력해주세요 : ') #2. 지역 이름 입력
a = df.index.str.contains(name) #3. 해당 행을 찾아서 해당 지역의 인구 구조를 저장
# str.contains(): 지정한 문자열이 포함되어 있는지 확인
# 검색 문자열이 포함되어 있는 경우에는 True를 반환
df2 = df[a]
df2
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
df2.T.plot()
plt.show()
```

궁금한 지역(A)의 인구 구조 비율과 차이 계산하기​

```python
import numpy as np
# 궁금한 지역 A의 인구 비율에서 B의 인구 비율을 뺀다 
x = df.sub(df2.iloc[0], axis = 1)
# A의 인구 비율에서 B의 인구 비율을 뺀 값의 제곱을 모두 더한다.
y = np .power(x,2)
z = y.sum(axis =1)
```

가장 차이가 적은 상위 5개 지역 시각화하기

```python
# 그 차이가 가장 작은 지역 5곳 찾는다
i = z.sort_values().index[:5]
#print(i)
#결과를 걱은 선 그래프로 보여준다
df.loc[i].T.plot()
plt.show()
```

전체 코드

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
df = pd.read_csv(directory_path + 'age_v2.csv', encoding='cp949', index_col = 0)
df = df.div(df['총인구수'], axis = 0)     # ①-1. 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수']  # ①-2. 총인구수, 연령구간인구수 열 삭제
name = input('원하는 지역의 이름을 입력해주세요 : ')  # ② 지역 이름 입력
a = df.index.str.contains(name)  # ③ 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2 = df[a]
df.loc[np.power(df.sub(df2.iloc[0], axis = 1), 2).sum(axis = 1).sort_values().index[:5]].T.plot()
plt.show()
```
