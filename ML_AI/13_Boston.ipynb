# 15장 선형 회귀 적용하기
1978년, 집값에 가장 큰 영향을 미치는 것이 ‘깨끗한 공기’라는 연구 결과가 하버드대학교 도시개발학과에서 발표됨.
이들은 자신의 주장을 뒷받침하기 위해 집값의 변동에 영향을 미치는 여러 가지 요인을 모아서 환경과 집값의 변동을 보여주는 데이터셋을 만듦.
그로부터 수십 년 후, 이 데이터셋은 머신러닝의 선형 회귀를 테스트하는 가장 유명한 데이터로 쓰이고 있음.

```python
#-*- coding: utf-8 -*-
#from keras.models import Sequential
from tensorflow.keras.models import Sequential
#from keras.layers.core import Dense
from tensorflow.keras.layers import Dense

from sklearn.model_selection import train_test_split

import numpy
import pandas as pd
import tensorflow as tf

# seed 값 설정
seed = 0
numpy.random.seed(seed)
# tf.random.set_seed(3) # commendted out by hsp
# tf.random.set_random_seed(3)

df = pd.read_csv("/Users/crunky/Desktop/인딥개/12주차/dataset/housing.csv", delim_whitespace=True, header=None)
```

```python
dataset = df.values
X = dataset[:,0:13]
Y = dataset[:,13]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam')
```

```python
model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
# 예측 값과 실제 값의 비교
Y_prediction = model.predict(X_test).flatten()
for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print("실제가격: {:.3f}, 예상가격: {:.3f}".format(label, prediction))
```
